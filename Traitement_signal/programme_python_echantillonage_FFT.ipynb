## Fréquence d'échantillonage

```python
from pylab import *

# parametres
fs=1000;
Ne=1024;
Tv=2/fs;
```

```python
#fréquence d'échantillonage
#fe=fs;
fe=2.1*fs;
#fe=16/17*fs;
#fe=16*fs; 
```

```python
te=linspace(0,Ne/fe,Ne) # les instants d’echantillonnage

def s(t):return (sin(2*pi*fs*t)) # le signal à étudie

#création du spectre 
spectre=(fft(s(te))/(Ne/2));
f=fftfreq(Ne,1/fe)

#création du graphique
t=linspace(0,Tv,1000)
subplot(211);plot(t,s(t),lw=2,label ='s(t)')
plot(te,s(te),'ro',label ='s(k/fe)')
axis([0,Tv,-1.1,1.1]);grid();legend() # cadrage
subplot(212);plot(f,abs(spectre),label='spectre')
axis([0,max(f),0,1.2]);legend();grid()
show()

#détermination de la fréquence correspondant au pic du spéctre:
print('fréquence du signal :',f[abs(spectre).argmax()], 'Hz')
```
